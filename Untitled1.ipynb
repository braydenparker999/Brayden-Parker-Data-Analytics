```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/braydenparker/Downloads/companies.csv')
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='state', data=df)
plt.show()
```

```python
plt.figure(figsize=(30, 6))
sns.barplot(x='Industry', y='job_openings', data=df)
plt.show()
```

```python
df_agg = df.groupby('founded')['growth_percentage'].mean().reset_index() 
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x='founded', y='growth_percentage', data=df_agg)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x='employees', y='estimated_revenues', data=df)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.histplot(x='founded', data=df)
plt.show()
```
