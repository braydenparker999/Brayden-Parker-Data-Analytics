In this analysis we'll be exploring the trends in housing in Melbourne, Australia and how relations and correlations found in the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/braydenparker/Brayden-Parker-Data-Analytics/housing.csv')
df.head()
```

Let's try and find a correlation between property price and landsize

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x='Landsize', y='Price', hue='Type', data=df)
plt.show()
```

It's hard to see a correlation between price and landsize here

Now let's look for average price per year built of these homes.

```python
df_agg_avgpriceyear = df.groupby('YearBuilt')['Price'].mean()
df['df_agg_avgpriceyear'] = df_agg_avgpriceyear
df_agg_avgpriceyear.head()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x='YearBuilt', y='Price', data = df_agg_avgpriceyear.reset_index())
plt.show()
```

There's little correlation here, but the general trend looks to be decreasing in price as the years go on.

Now we will look into the average price per region of the homes.

```python
df_agg_avgpriceregion = df.groupby('Regionname')['Price'].mean()
df_agg_avgpriceregion.head()
```

```python
plt.figure(figsize=(20,  10))
sns.barplot(x='Regionname', y='Price', data = df_agg_avgpriceregion.reset_index())
plt.show()
```

We can clearly see the different average prices per region of homes here.

Now we will look at a histogram of property prices.

```python
plt.figure(figsize=(10, 10))
sns.histplot(x='Price', data = df)
plt.show()
```

We can observe a left skew in this data.

Now we will look at a boxplot of the years that each property was built.

```python
plt.figure(figsize=(20, 20))
sns.boxplot (x='YearBuilt', y='Regionname', data = df)
plt.show()
```
