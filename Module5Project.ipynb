In this analysis we will analyze the various relationships in the data we have and try to find correlation between different values.

```python
import pandas as pd
df = pd.read_csv('wine.csv')
df.head()
```

```python
print(df.shape)
```

# Question 1
In this section we will find which country, on average, produces the highest quality wine/the most points.

```python
Avg_points_by_country = df.groupby('country')['points'].mean()
Avg_points_by_country.sort_values(ascending=False)
```

As we can see from this data, England on average, produces the wine with the most points.

# Question 2
Now we will find which taster on average gives the lowest score. Through this we can see those that usually give lower scores, that the wines they give higher scores could be more justified, because they are not partial to always giving high scores.

```python
taster_mean_points = df.groupby('taster_name')['points'].mean()
taster_mean_points.sort_values(ascending=True)
```

Alexander Peartree usually gives the lowest score.

# Question 3
Next we will see what the average price of wine is for each variety.

```python
mean_price_by_variety = df.groupby('variety')['price'].mean()
mean_price_by_variety.sort_values(ascending=False)
```

Ramisco is by far the most expensive bottle, almost averaging $500 per bottle.

# Question 4
Here we are analyzing the wine by year, and finding which year averaged the most points.

```python
wine_year = df['title'].str.extract(r'(\d{4})')
df['year'] = wine_year
wine_year_mean_points = df.groupby(wine_year.squeeze())[['points']].mean()
wine_year_mean_points.sort_values(by='points', ascending=False)
```

# Question 5, 6 & 7
In the next section we analyze the specific contents of the description to see if there's a correlation between containing the word "depth" and getting more points on average.

```python
depth = df['description'].str.contains('depth')
df['contains_depth'] = depth
contains_depth_mean_points = df.groupby('contains_depth')['points'].mean()
contains_depth_mean_points.sort_values(ascending=False)
```

We find that it is true, on average descriptions containing the word "depth" get more points.

Next we will be doing the exact same thing but for the word "fruity"

```python
fruity = df['description'].str.contains('fruity')
df['contains_fruity'] = fruity
contains_fruity_mean_points = df.groupby('contains_fruity')['points'].mean()
contains_fruity_mean_points.sort_values(ascending=False)
```

It appears the the word "fruity" has the opposity correlation of depth and reviews containing it average less points. Although it can definitely be argued that the differences on both "depth" and "fruity" are so slim that it is a weak correlation.

Next, we will do the same thing for the word "herbal"

```python
herbal = df['description'].str.contains('herbal')
df['contains_herbal'] = herbal
contains_herbal_mean_points = df.groupby('contains_herbal')['points'].mean()
contains_herbal_mean_points.sort_values(ascending=False)
```

This also is a very slim difference, but reviews containing the word "herbal" tend to get less points on average.

# Question 8
Next we will find the correlation between length of review in characters, and the points of the review

```python
number_of_characters = df['description'].str.len()
df['number_of_characters'] = number_of_characters
df['number_of_characters'].corr(df['points'])
```

There is a moderate postive relationship

Here we are finding the average points for each region that the wine comes from.

```python
region = df['title'].str.extract(r'\((.+)\)')
df['region'] = region
region_mean_points = df.groupby('region')['points'].mean()
region_mean_points.sort_values(ascending=False)
```

# Question 9
Next we will find which region within in Sicily & Sardinia produces the best wine, on average.

```python
sicily_sardinia = df[df['province'] == 'Sicily & Sardinia']
sicily_sardinia['region'] = sicily_sardinia['title'].str.extract(r'\((.+)\)')
region_mean_points = sicily_sardinia.groupby('region')['points'].mean()
region_mean_points.sort_values(ascending=False)
```

It appears that Faro produces the best wine on average.
